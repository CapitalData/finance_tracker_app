## Mermaid diagramming settup
* Inv chain profiler,
`inv_chain_test(inv_pay_df)` 

    tracks resource flow, via invoices names
    Job: {Client} X/> {inv_nm} X/> {ACD} X/> {inv_nm} X/> {worker}

    example 
    Job: {Axcel} X{4}X {ACD} X{Rlee_2}> {R.Lee}
    * Incomplete, Raymond has invoiced ACD, ACD has not paid or invoiced Axcel
    
    Job: {Axcel} > {4} > {ACD} > {Rlee_2} > {R.Lee}
    * Complete, Ray has invoced and been paid ACD as invoiced and been paid
    
    legend 
    > complete
    X incomplete
    {} named entity

```python
import acd_datatool as acd
```

```python
acd.mm("""
classDiagram
    %%note for invoice "invoicing and payments"

    class accountable{
        +name
        +status 
        +amount
        +from
        +to
        +date_initiated
        +date_resolved
    }

    class invoice{
        +job_ls[]
        }

    class job{
        +service_dates
        +staff[]
        }

    style accountable fill:#f9f,stroke:#333,stroke-width:4px    
    style invoice fill:#f9f,stroke:#333,stroke-width:4px    
    style job fill:#f9f,stroke:#333,stroke-width:4px    

    class entity{
        
        -make_inovice(invoice, job, client)
        -pay_invoice(invoice, job)
        -receive_pay()
        }

    class talent_co{
        -staff(worker, job)
    }
    class worker{
        +is_fully_paid
        +accept_job()
        +work_job()
    }
    class edu_cont_co{
        +receive_job_request()
        +_job_request()
    }
    class client_co{
        +request_job()
    }


    accountable--|>invoice
    accountable--|>job

    entity--|>worker
    entity--|>edu_cont_co
    entity--|>client_co
    entity--|>talent_co
""")
```

```python
acd.mm("""
classDiagram
direction RL

        class job{
        +Name
        +status
        +tasks=[]
    }

        class invoice{
        +Name
        +status
        +Jobs=[]
        +amounts=[]
        +hours=[]
    }

    class worker{
        +Name
        +Jobs=[]
        +is_fully_paid
        +receive_job()
        +accept_job()
        +complete_job()
        +Make_invoice()
    }

    class talent_co{
        +Name
        +Inv_Wk=[]
        +Inv_Brkr=[]
        +Jobs=[]
        +receive_job()
        +staff_job(worker, job)
        +Pay_invoice()
        +Make_invoice() 

    }

    class edu_cont_co{
        +Name
        +Inv_Brkr=[]
        +receive_job_request()
        +request_job()
        %%+bill_invoice() 
        +pay_invoice()
    }
    
    class client_co{
        +Name
        +Make_job()
        %%+pay_invoice()
    }

    talent_co-->edu_cont_co :Bill Brk_Invoice
    worker-->talent_co :Bill Wk_Invoice

    talent_co<--edu_cont_co :Pay Brk_Invoice
    worker<--talent_co :Pay Wkr_Invoice

    talent_co<--edu_cont_co :Job Ask
    edu_cont_co<--client_co :Job Ask
    worker<--talent_co :Job Ask

    %%edu_cont_co-->client_co :Bill Invoice
    %%edu_cont_co<--client_co :Pay Invoice

    style client_co fill:#f9f,stroke:#333,stroke-width:4px 
    style job fill:#118f99,stroke:#333,stroke-width:4px
    style invoice fill:#118f99,stroke:#333,stroke-width:4px 
""")

"{client}-job->{edu_cont_co}-job->{talent_co}<-invoice/job->{worker}"
```
