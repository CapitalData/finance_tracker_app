```python

def load_method(method, secrets_ls):

    """## local implies that you have the google sheets ssh key file on your local machine. 
    ## colab assumes that you are using colab with colab secrets
    # e.g. for secrets API locally
    # """

    #Method 1 -  this loads the ssh key for the service account
    if method == 'l' or 'L' or 'local':
  
        #For colab use secrets
        ## *** get API code from secrets or use server side key file

        # api secrets file
        with open (secrets_ls[1], 'r') as f:
            api_dict= json.loads(f.read())

        # google secrets account Path to the credentials file
        filename = secrets_ls[0]
        
        # Define the scope
        scope = ['https://spreadsheets.google.com/feeds', 
            'https://www.googleapis.com/auth/drive']

        # Load credentials from the file
        credentials = ServiceAccountCredentials.from_json_keyfile_name(filename, scope)

        g_conect = gspread.authorize(credentials)

    #Method 2 use on colab - this injects the credential from the user authentication
    elif method == 'c' or 'C' or 'colab_auth':

        ### this will step you thourgh an authentication routine thorugh popup windows
        from google.colab import auth
        auth.authenticate_user()

        from google.auth import default

        creds, _ = default()
        g_conect = gspread.authorize(creds)

        from google.colab import drive
        drive.mount('/content/gdrive')
        # Load the Google Sheets
        # gc = gspread.service_account(filename=filename)  # Replace with your credentials file

    else:
        print("no valid method selected")
        sys.exit()

    return (g_conect, api_dict)
    
```

```python
## testing suite ___ DATA NEEDS TO BE LOADED
 # - incosistencies arising from double (or more) entry discrepancies tested and logged

####### invoice/ job double entries
def inv_dup_test(inv_pay_df):
    """ # 1- each invoice name should be unique """    
    #for inv in inv_pay_df['inv_name'] :
    inv_nm_ct=inv_pay_df.inv_name.value_counts()
    repeated_inv_names = inv_nm_ct[inv_nm_ct>1]
    if len(repeated_inv_names) > 0:
        print('DATA ISSUE!: repeated invoice names:', repeated_inv_names)



# 2- Each invoice name should be listed in the job sheet"""
# there seems to be an issue recognizing numeric invoice names. they come in from googlesheets as NAN?
#jobsheet_inv_nm = 

if verbose == True:
    print(jobs_df['ACD invoice'].value_counts())
    print(jobs_df['Worker invoice'].unique())
    #print(jobsheet_inv_nm)
    print(jobs_df['ACD invoice'])

    display(inv_pay_df.columns)
    display(jobs_df.columns)
    display(inv_pay_df.head())
    display(jobs_df.head())


#- The invoice total shoud be the same as the sum of the jobs listed for that invoice
####### violations of data ontology, state and maturity expectaions tested and logged


def inv_stat_tbl(jobs_df):
    """# generate a counts dataframe for invoice status"""
    df1=jobs_df['ACD_inv_status'].value_counts()
    df2=jobs_df['Wk_inv_status'].value_counts()

    col_nms=['acd_inv','wkr_inv']

    df1=df1.reset_index()
    df1.columns=['item', col_nms[0]]
    df2=df2.reset_index()
    df2.columns=['item', col_nms[1]]

    df1.head()
    inv_stat_count=pd.merge(df1, df2, on='item', how='outer').fillna(0)
    
    inv_stat_count[col_nms[0]]=inv_stat_count[col_nms[0]].astype('int')
    inv_stat_count[col_nms[1]]=inv_stat_count[col_nms[1]].astype('int')
    return(inv_stat_count)


def inv_chain_test(inv_df=inv_pay_df, jb_df=jobs_df, verbose==False):
    """tracks resource flow, via invoices names
    Job: {Client} X/> {inv_nm} X/> {ACD} X/> {inv_nm} X/> {worker}

    example 
    Job: {Axcel} X{4}X {ACD} X{Rlee_2}> {R.Lee}
    * Incomplete, Raymond has invoiced ACD, ACD has not paid or invoiced Axcel
    
    Job: {Axcel} > {4} > {ACD} > {Rlee_2} > {R.Lee}
    * Complete, Ray has invoced and been paid ACD as invoiced and been paid
    
    legend 
    > complete
    X incomplete
    {} named entity"""

    inv_df=inv_pay_df
    jb_df=jobs_df

    new_df=jb_df[['job','direct client', 
    'ACD invoice', 'Worker invoice','Teacher',
    'ACD_inv_status', 'Wk_inv_status']]

    ## chart the transition status
    status=['</X','</X','</X','</X']

    # Iterate over each row and print it
    for index, row in new_df.iterrows():
        #print(f"Row {index}: {row.to_dict()}")

        ## chart the entities
        job=row.iloc[0]
        client=row.iloc[1]
        inv1_nm=row.iloc[2]
        home_co='ACD'
        inv2_nm=row.iloc[3]
        worker=row.iloc[4]

        # update the statuses from the dict
        respon_dct={'unsubmitted':['X','X'], 
                    'submitted':['X','<'], 
                    'paid':['<','<']}
        
        if row['ACD_inv_status'] in respon_dct.keys():
            status[0:2]=respon_dct[row['ACD_inv_status']]
            #print(f'{}, {status}')
            print(status)
        
        if row['Wk_inv_status'] in respon_dct.keys():
            status[2:4]=respon_dct[row['Wk_inv_status']]
    
        respon_dct[row['ACD_inv_status']]=status[0:2]

    # report out
        if verbose==True 
            print(f"ACD_inv_status: {row['ACD_inv_status']} Wk_inv_status: {row['Wk_inv_status']}")
            print(status)
        print(f'{job} Job: {client} {status[0]} {inv1_nm} {status[1]} {home_co} {status[2]} {inv2_nm} {status[3]} {worker}')
        print('-'*50)

def run_test(funct, *args, **kwargs):
    """Utility function for running a test and reporting the test name """
    
    print(f'\n\n running test:: {funct.__name__}')
    
    return funct(*args, **kwargs)
    
```
